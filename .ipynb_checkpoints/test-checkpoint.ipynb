```python
from pycoingecko import CoinGeckoAPI
import pandas as pd
import requests
import plotly.express as px
from datetime import timezone
from datetime import datetime
pd.options.display.float_format = '{:.2f}'.format
```

```python
#initialize the Cg API instance
cg = CoinGeckoAPI()
```

```python
url="https://api.coingecko.com/api/v3/coins/markets?vs_currency=usd&order=market_cap_desc"
data=requests.get(url).json()
```

```python
test_df=pd.DataFrame.from_dict({'id':[item['id'] for item in data],  'market_cap'  :[item['market_cap'] for item in data]}).sort_values(by=['market_cap'],ascending=False).head(250)
test_df['rank']=test_df['market_cap'].rank(ascending=False).astype(int)
test_df
```
